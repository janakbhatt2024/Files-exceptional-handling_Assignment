Q2 What is exception handling in Python?

Answer:


Exception handling in Python is a mechanism that allows you to manage errors or exceptional conditions that may occur during the execution of a program. Instead of the program crashing when an error occurs, Python provides a way to gracefully handle these errors using the try, except, else, and finally blocks.

try: Try to execute the code that might raise an exception.

except: Handle specific exceptions (errors).

else: Execute if no exception occurs.

finally: Always execute, even if an exception occurs, typically for cleanup tasks.

By using exception handling, you can make your programs more robust and avoid unexpected crashes by managing errors gracefully.

Q3 What is the purpose of the finally block in exception handling?

Answer:

Purpose of the finally Block:


Guaranteed Execution: The code inside the finally block will always run, no matter what happens in the try or except blocks. This ensures that any necessary cleanup or resource management tasks (such as closing files, releasing locks, or cleaning up database connections) are always performed.

Post-Error Cleanup: Even if an exception is raised and handled, or if no exception occurs, the finally block ensures that essential actions (like closing a file or freeing resources) happen at the end.

Q15 How does Python's garbage collection system work?

Answer:

Pythons garbage collection system is a combination of reference counting and cyclic garbage collection:

Reference counting: Tracks the number of references to an object, and when it reaches zero, the object is deallocated.

Cyclic garbage collection: Detects and collects objects involved in cyclic references, which would not be collected by reference counting alone.

Generational collection: Objects are organized into generations based on their age, with younger objects being collected more frequently than older ones.

gc module: Provides the ability to interact with and control the garbage collection process manually.

Q23 What are the challenges associated with memory management in Python?

Memory management challenges in Python include:

Garbage Collection Issues: Circular references can prevent memory deallocation, and garbage collection timing is unpredictable.

Memory Leaks: Unintended references or improper resource management can cause leaks.

Fragmentation: Memory fragmentation can hinder efficient allocation of large memory blocks.

Inefficient Memory Use: Internal overhead of data structures and unnecessary object copies can waste memory.

Immutable Objects: Interned immutable objects may increase memory usage.

C Extensions: Third-party C extensions may cause memory management issues.

```python
# Python script to read from one file and write to another

# Specify the input and output file paths
input_file = 'input.txt'
output_file = 'output.txt'

# Open the input file in read mode and the output file in write mode
try:
    with open(input_file, 'r') as infile:
        content = infile.read()  # Read the entire content of the input file

    with open(output_file, 'w') as outfile:
        outfile.write(content)  # Write the content to the output file

    print(f"Content from {input_file} has been written to {output_file}")
except FileNotFoundError:
    print(f"Error: The file {input_file} does not exist.")
except Exception as e:
    print(f"An error occurred: {e}")
```

Q6 Write a Python program that logs an error message to a log file when a division by zero exception occurs.

```python
import logging

# Set up logging configuration
logging.basicConfig(
    filename='error_log.txt',  # Log file name
    level=logging.ERROR,       # Log level
    format='%(asctime)s - %(levelname)s - %(message)s'  # Log message format
)

def divide_numbers(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError as e:
        # Log the error message
        logging.error(f"Error: Division by zero. a = {a}, b = {b}. Exception: {e}")
        return None

# Example usage
num1 = 10
num2 = 0

# Attempt to divide
result = divide_numbers(num1, num2)

if result is None:
    print("Error occurred, check the log file.")
else:
    print(f"Result: {result}")
```

Q8 Write a program to handle a file opening error using exception handling.

Answer:

```python
try:
    # Attempt to open a file that may not exist
    with open('example.txt', 'r') as file:
        content = file.read()
        print("File content:")
        print(content)

except FileNotFoundError:
    # Handle the case where the file is not found
    print("Error: The file 'example.txt' was not found.")

except IOError:
    # Handle any I/O related errors
    print("Error: An I/O error occurred while trying to open the file.")

except Exception as e:
    # Catch any other unexpected errors
    print(f"An unexpected error occurred: {e}")

else:
    # This block runs if no exception was raised
    print("File opened and read successfully.")

finally:
    # This block always runs, regardless of whether an exception occurred or not
    print("Execution completed.")
```

Q12 Write a program that demonstrates using multiple except blocks to handle different types of exceptions.

Answer

```python
def divide_numbers(a, b):
    try:
        # Trying to divide two numbers
        result = a / b
        print(f"The result of {a} divided by {b} is: {result}")

    # Handling division by zero error
    except ZeroDivisionError:
        print("Error: You can't divide by zero!")

    # Handling type errors (e.g., if inputs are not numbers)
    except TypeError:
        print("Error: Both inputs must be numbers!")

    # Handling value errors (e.g., when an invalid value is passed)
    except ValueError:
        print("Error: Invalid value provided!")

    # Handling all other exceptions (general exception handler)
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

    else:
        print("Division was successful!")

    finally:
        print("Execution complete.")

# Test cases
divide_numbers(10, 5)       # Normal case, no error
divide_numbers(10, 0)       # Division by zero error
divide_numbers(10, 'a')     # Type error (invalid data type)
divide_numbers('ten', 5)    # Type error (invalid data type)
```

Q18 How would you implement a basic logging setup that logs to a file with rotation after 1MB?

Answer:

```python
import logging
from logging.handlers import RotatingFileHandler

# Set up the logger
logger = logging.getLogger('my_logger')
logger.setLevel(logging.DEBUG)  # Set the logging level (e.g., DEBUG, INFO)

# Set up a rotating file handler
log_file = 'my_log.log'
max_log_size = 1 * 1024 * 1024  # 1MB in bytes
backup_count = 3  # Keep 3 backup log files

handler = RotatingFileHandler(log_file, maxBytes=max_log_size, backupCount=backup_count)
handler.setLevel(logging.DEBUG)

# Set up a formatter
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example logging statements
logger.debug("This is a debug message")
logger.info("This is an info message")
logger.warning("This is a warning message")
logger.error("This is an error message")
logger.critical("This is a critical message")
```

Q21 Write a Python program that reads a file and prints the number of occurrences of a specific word.

Answer:

```python
def count_word_in_file(file_path, word):
    # Open the file in read mode
    try:
        with open(file_path, 'r') as file:
            # Read the entire content of the file
            content = file.read()

            # Convert the content to lowercase to make the search case-insensitive
            content = content.lower()

            # Count occurrences of the word in the content
            word_count = content.split().count(word.lower())  # Ensure word is case-insensitive

            return word_count
    except FileNotFoundError:
        print(f"File at {file_path} not found.")
        return None

# Example usage
file_path = input("Enter the file path: ")
word = input("Enter the word to search for: ")

word_count = count_word_in_file(file_path, word)
if word_count is not None:
    print(f"The word '{word}' occurred {word_count} times in the file.")
```

Q22 How can you check if a file is empty before attempting to read its contents?

Answer:

```python
import os

def count_word_in_file(file_path, word):
    if not os.path.exists(file_path) or os.stat(file_path).st_size == 0:
        return 0  # File doesn't exist or is empty

    with open(file_path, 'r') as file:
        content = file.read().lower()
        return content.split().count(word.lower())

# Example usage
file_path = input("Enter the file path: ")
word = input("Enter the word to search for: ")

word_count = count_word_in_file(file_path, word)
print(f"The word '{word}' occurred {word_count} times in the file.")
```
